```python
# 労働時間が9時間以上のデータを抽出
df = pd.read_excel('/content/masuda.xlsx', sheet_name='202301')
df_overwork = df.query('労働時間 >= 9')
print(df_overwork)
```

```python
# 1月から4月までの労働時間が9時間以上のデータを抽出
all_df = pd.read_excel('/content/masuda.xlsx', sheet_name=None)
overwork_list = []
for sheet in all_df.keys():
    df_overwork = all_df[sheet].query('労働時間 >= 9')
    overwork_list.append(df_overwork)

# concatはデータフレームを結合する関数
df = pd.concat(overwork_list)
print(df)
```
