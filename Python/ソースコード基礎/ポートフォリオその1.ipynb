## if ~ else
比較演算子の結果によって、処理を変えることができる

```
if 条件:
    条件が true の時に実行するもの
else:
    条件が false の時に実行するもの
```

````python
a = 49

if a >= 50:
    print('aは50以上である')
else:
    print('50以下')
````

````python
# aに100という値を代入
a = 30

# a > 50 が 正しければ、以下の処理を実行
if a > 50:
  print('50よりも大きい')

# a > 50 が正しくなければ、以下の処理を実行
else:
  print('50以下')
````


- 例えば、ECサイトを運営しているとして、「3回以上購入したことがあるユーザーか、そうでないかで送付するクーポンの種類を分ける」など、条件に応じて処理を分けるというのは多く行われる


### インデント

- Pythonでは、処理のひと固まりを、行頭の空白で表します
- 上記では、`print('50よりも大きい')`や`print('50以下')`の前に4つの半角スペースがあります
- 通常、半角スペース4つ（もしくは2つ）を使います
- このような行頭の空白を**インデント**と呼ぶ
- オフサイドルールと言う

````python
a = 100

if a > 50:
    print('50よりも大きい1')
    print('50よりも大きい2')
    print('50よりも大きい3')
else:
    print('50以下1')
    print('50以下2')
````


インデントではなく、`{}` などで囲う言語も

Python では、上記のように関数を定義する
最初の行は `def` で始まり、関数内で使用する変数を引数と呼ぶ。

Python
    def 関数名(引数, ...):



`:` 以降で、インデント入れて、関数内の処理を記述。
インデントが通常に戻るまでは、関数内の処理を行う。




````python
# 関数と引数。FizzBuzzを実行する関数を作成
def fizzbuzz():
    x = 3
    if x % 15 == 0:
        result = 'FizzBuzz'
    elif x % 3 == 0:
        result = 'Fizz'
    elif x % 5 ==0:
        result = 'Buzz'
    else:
        result = x
    return result

# 関数の実行結果をrに入れて実行
r = fizzbuzz()
print(r)
````

````python
# ただ、上記の場合だと、x = 3のときのみしか使えない関数となるので、引数でこれを解決する

def fizzbuzz(x):
    if x % 15 == 0:
        result = 'FizzBuzz'
    elif x % 3 == 0:
        result = 'Fizz'
    elif x % 5 ==0:
        result = 'Buzz'
    else:
        result = x
    return result

# この3という数字がdef fizzbuzz(x)のx部分の値となる
r = fizzbuzz(5)
print(r)
````

````python
# 引数=3の実行結果を確認
r1 = fizzbuzz(3)

# 引数=5の実行結果を確認
r2 = fizzbuzz(5)

# 引数=15の実行結果を確認
r3 = fizzbuzz(15)

# 引数=100の実行結果を確認
r4 = fizzbuzz(97)

# 全ての結果を出力
print(r1, r2, r3, r4)
````

## ループ処理

* きまった操作の繰り返しを行う際は、**for文**を用います。
* 「for **変数名** ~ in **データの集まり**」といった形で用いることができます



````python
num_list = [1, 2, 3, 4, 5]
````

````python
n = 0
for i in num_list:
    print(i)
    print(i * 10) # print(i * 10)を実行
    print(i + 800)
    print('------')

print('ループ終了')
````

 上記のfor文では、2行目で、num_list から一つずつ要素を取り出して変数 `i`に代入しています
 続いて、3~5行目で、取り出した要素に対して何かしら演算をして表示しています

最後の、`print('ループ終了')`はインデントの外なので、最後に一度だけ実行される

````python
# これでいちいち出力だと時間がかかる
i = 0
print(i)
i = 1
print(i)
````
