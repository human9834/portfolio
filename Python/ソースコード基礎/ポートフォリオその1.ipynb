## if ~ else
比較演算子の結果によって、処理を変えることができる

```
if 条件:
    条件が true の時に実行するもの
else:
    条件が false の時に実行するもの
```

````python
a = 49

if a >= 50:
    print('aは50以上である')
else:
    print('50以下')
````

````python
# aに100という値を代入
a = 30

# a > 50 が 正しければ、以下の処理を実行
if a > 50:
  print('50よりも大きい')

# a > 50 が正しくなければ、以下の処理を実行
else:
  print('50以下')
````


- 例えば、ECサイトを運営しているとして、「3回以上購入したことがあるユーザーか、そうでないかで送付するクーポンの種類を分ける」など、条件に応じて処理を分けるというのは多く行われる


### インデント

- Pythonでは、処理のひと固まりを、行頭の空白で表します
- 上記では、`print('50よりも大きい')`や`print('50以下')`の前に4つの半角スペースがあります
- 通常、半角スペース4つ（もしくは2つ）を使います
- このような行頭の空白を**インデント**と呼ぶ
- オフサイドルールと言う

````python
a = 100

if a > 50:
    print('50よりも大きい1')
    print('50よりも大きい2')
    print('50よりも大きい3')
else:
    print('50以下1')
    print('50以下2')
````


インデントではなく、`{}` などで囲う言語も

### if ~ elif ~ else




````python
a = 1

if a > 50:
    print('50より大きい')
elif a > 30:
    print('30より大きく50以下')
else:
    print('30以下')
````

````python
# aに値を代入
a = 60

if a > 50: # a > 50 が Trueのときに以下の処理を実行
    print('50より大きい') 
elif a > 30: # a > 50 は Falseだが、a > 30 は Trueのときに以下の処理を実行
    print('30より大きく50以下') 
else: # a > 50 は Falseで、また a > 30も Falseのときに以下の処理を実行
    print('30以下')    
````

````python
a = 9
b = 2
c = a * b

if c > 50:
    print('50よりも大きい')
elif c > 30:
    print('30よりも大きく50以下')
elif c > 20:
    print('20よりも大きく30以下')
if c > 10:
    print('10よりも大きく20以下')
else:
    print('10以下')               
````

````python
a = "red"

if a == 'red':
    print('apple')
elif a == 'blue':
    print('sky')
else:
    print('earth')
````

## 関数

- 上で登場したif 文の例のように、同じような処理を何度も書くのは面倒
- 処理を**関数**という形で定義することで、何度も再入力せずに実行できて便利
- エクセルの、sum(), min(), max() も関数
- よく用いる計算や処理をひとまとめにして、オリジナルの関数を作ることができる

````python
a = 80

if a > 50:
    print('50よりも大きい')
else:
    print('50以下')
````

````python
def is_bigger_50(numer):
    if numer > 50:
        print('50よりも大きい')
    else:
        print('50以下')
````

````python
is_bigger_50(30)
````

````python
is_bigger_50(100)
````

Python では、上記のように関数を定義する
最初の行は **`def`** で始まり、関数の中で使用する変数を列挙します。
そのような、関数内で使用する変数を **引数**と呼びます。

------
```Python
    def 関数名(引数, ...):
```
------


`:` 以降で、インデント入れて、関数内の処理を記述。
インデントが通常に戻るまでは、関数内の処理をみなされます。




````python
# 徒歩の消費カロリー
# 3 × 体重 × 運動時間(分数) × 1.05 ÷ 60
weight = 60
minutes = 50
3 * weight * minutes * 1.05 / 60
````

````python
# 体重と運動時間（分数）を与えると結果が戻ってくる

def calories_walking(weight,minutes):
    return 3 * weight * minutes / 60 * 1.05
````

````python
x = calories_walking(60, 30)
print(x)
````

````python
calories_walking(70, 90)
````

- 関数内に、`return`という文があります。
- これは、returnの後ろに続く値を関数の呼び出し元に返すということを意味しています
- この、返された値を、そのままですが、**返値**と言います

## ループ処理

* きまった操作の繰り返しを行う際は、**for文**を用います。
* 「for **変数名** ~ in **データの集まり**」といった形で用いることができます



````python
num_list = [1, 2, 3, 4, 5]
````

````python
n = 0
for i in num_list:
    print(i)
    print(i * 10) # print(i * 10)を実行
    print(i + 800)
    print('------')

print('ループ終了')
````

 上記のfor文では、2行目で、num_list から一つずつ要素を取り出して変数 `i`に代入しています
 続いて、3~5行目で、取り出した要素に対して何かしら演算をして表示しています

最後の、`print('ループ終了')`はインデントの外なので、最後に一度だけ実行される

````python
# これでいちいち出力だと時間がかかる
i = 0
print(i)
i = 1
print(i)
````

````python
# なのでfor文で効率化(rangeはPythonで連続する整数のリストを自動で生成する関数)
for i in range(5):
    print(i)
````

````python
# 1から始めたい場合(0)
for i in range(1, 5):
    print(i)
````

````python
# 1～9で3つ飛ばしにする

for i in range(1, 9, 3):
    print(i)
````

````python
# 1～4で2つ飛ばしにする(末の数字が飛ばす数)

for i in range(1, 5, 2):
    print(i)
````

 変数wordをfor文×range()で出力する

````python
word = 'Python'

for i in range(6):
    print(word[i])
````

````python
# lengthのlen()で文字列の長さを指定する

for i in range(len(word)):
    print(word[i])
````

````python
# 簡略化(理由が無い限りはこれでいい)

for s in word:
    print(s)
````
