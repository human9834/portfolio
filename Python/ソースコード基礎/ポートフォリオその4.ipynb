### ライブラリ

- ライブラリとは、再利用しやすいように複数の関数をまとめたもの
- まずは、mathという様々な数学関数を呼び出せるライブラリを使用

```python
import math
```

```python
# 平方根
math.sqrt(3)
```

```python
# モジュールに含まれる関数やクラス、定数などに関する説明や使用方法が表示される
help(math)
```

#### math.関数名のリスト

- math.sin(x) : サイン関数
- math.cos(x) : コサイン関数
- math.factorial(x) : 階乗
- math.sqrt(x): 平方根

```python
# sin. cos
math.sin(10)
```

```python
math.cos(2)
```

```python
# 3! = 3 × 2 × 1, 階乗
math.factorial(3)
```

```python
# 平方根
math.sqrt(10)
```

#### ライブラリの読み込みを短縮する

```python
# 'm'という名前でライブラリを使用
import math as m
m.sqrt(3)
```
