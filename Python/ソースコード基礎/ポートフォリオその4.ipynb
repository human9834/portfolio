### ライブラリ

- ライブラリとは、再利用しやすいように複数の関数をまとめたもの
- まずは、mathという様々な数学関数を呼び出せるライブラリを使用

```python
import math
```

```python
# 平方根
math.sqrt(3)
```

```python
# モジュールに含まれる関数やクラス、定数などに関する説明や使用方法が表示される
help(math)
```

#### math.関数名のリスト

- math.sin(x) : サイン関数
- math.cos(x) : コサイン関数
- math.factorial(x) : 階乗
- math.sqrt(x): 平方根

```python
# sin. cos
math.sin(10)
```

```python
math.cos(2)
```

```python
# 3! = 3 × 2 × 1, 階乗
math.factorial(3)
```

```python
# 平方根
math.sqrt(10)
```

#### ライブラリの読み込みを短縮する

```python
# 'm'という名前でライブラリを使用
import math as m
m.sqrt(3)
```

### 便利なライブラリ

- numpy: ベクトルや行列などの数値演算を行うことができる
- pandas: 表形式のデータを柔軟に操作・処理することができる、エクセル内でできることは代替可能
- matplotlib, seaborn, plotly: 柔軟な可視化を行うことができる

#### Numpy

```python
# asの後のnpは、省略形。今後npで使用できる
import numpy as np
```

```python
# 1次元配列の作り方
# listと近い
arr = np.array([1, 2, 3, 4])
arr
```

```python
# 2次元配列の作り方
# 今回は2行4列の配列を定義
arr2 = np.array([[1, 2, 3, 4], [3, 7, 8, 9]])
print(arr2)
```

```python
# .T で転置
arr2.T
```

```python
# 要素のアクセスは、インデックスを指定
arr2[1]
```

```python
# 例えばarr2[0][0]は1行目の1列目の要素を取得する事になる
arr2[1][2]
```

#### Pandas

```python
# エクセルでできるようなことは Pandas で代替可能
import pandas as pd
```

```python
# read_csv の結果を dfという変数に格納
# dfという変数に対して様々な処理が可能

df = pd.read_csv("/content/sample教材.csv")
df
```

```python
# numpyと同様にデータの次元を取得できる(20列、3行)
df.shape
```

```python
# 年齢列を抽出
df['年齢']
```

* 行番号や列番号を指定して抽出したい場合は、ilocを使う


```python
# indexが2の行（3行目）の抽出
df.iloc[2]
```

```python
# indexが2の行以降（3行目以降）の抽出
df.iloc[2:]
```

```python
# 2列目2行目をピックアップ。0、0で1列目1行目
df.iloc[1, 1]
```

```python
# 3行目～6行目の1列目～2列目を取得
df.iloc[2:6, 0:2]
```

*   describe()⇒各カラムの平均値や標準誤差、四分位数といった基礎統計量を求めることができる
*   to_csv(): csvファイル形式での保存ができる
*   リスト項目

```python
df['年齢'].mean()
```

```python
df['年齢'].sum()
```

```python
df.describe()
```

```python
# 新しい df2という変数に代入
df2 = df.describe()
df2
```

```python
# dfはデータフレーム
# データの保存
df2.to_csv('sample教材2.csv')
```
